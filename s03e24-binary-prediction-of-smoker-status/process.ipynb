# check null values

```python
def describe(df):
    desc = pd.DataFrame(index=df.columns)
    desc['type'] = df.dtypes
    desc['count'] = df.count()
    desc['null*'] = df.isnull().sum()
    desc['null%'] = desc['null*'] / desc['count'] * 100
    desc['unique*'] = df.nunique()
    desc['unique%'] = desc['unique*'] / desc['count'] * 100
    desc = pd.concat([desc, df.describe().drop(['count']).T], axis=1)
    display(desc)

describe(proc_train)
```
